```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
import json
from io import StringIO

data=[]

for i in range(1,41):
    url="https://www.carwale.com/"
    r=requests.get(url)
    soup=BeautifulSoup(r.text,"html.parser")
        
    car_tag=soup.findAll(name="div",attr={"class":"o-bWHzMb o-ducbvd o-cglRxs YONMcZ o-fpkJwH o-dCyDMp"})
        #data[]
    for car in car_tag:
            cardict={}
            
            car_names=car.find(name="div",attr={"class":"o-fznJDS o-ckGLSv o-fznJFI o-cMwvCl o-fznJzu o-vInIC o-fznJPk o-bOkGro o-efHQCX o-dEJOrr o-cpnuEd"})
            car_price=car.find(name="div",attr={"class":"o-Hyyko o-bPYcRG o-eqqVmt"})
            car_user_reviews=car.find(name="span",attr={"class":"o-eemiLE o-cYdrZi o-byFsZJ o-eNNlqh"})
            car_exepert_reviews=car.find(name="span",attr={"class":"o-eemiLE o-cYdrZi o-byFsZJ o-eNNlqh"})
            car_varients=car.find(name="title",attr={"class":"o-eZTujG o-jjpuv o-cVMLxW"})
            car_fuel_cc=car.find(name="span",attr={"class":"cpNAVm"})
            
            
            cardict["car_names"]=car_names.text.strip() if car_names else "N/A"
            cardict["car_price"]=car_price.text.strip() if car_price else "N/A"
            cardict["car_user_reviews"]=car_user_reviews.text.strip() if car_user_reviews else "N/A"
            cardict["car_expert_reviews"]=car_exepert_reviews.text.strip() if car_exepert_reviews else "N/A"
            cardict["car_varients"]=car_varients.text.strip() if car_varients else "N/A"
            cardict["car_fuel_cc"]=car_fuel_cc.text.strip() if car_fuel_cc else "N/A"
            
            data.append(cardict)

            json_data=json.dumps(data,indent=4)
            json_io=StringIO(data)
            df=pd.read_json(json_io)
            df.to_csv("car.csv",index=False)
            
            print(f"Saved{len(data)}Cars to final.csv")
```

```python
import pandas as pd
from bs4 import BeautifulSoup
import requests

AMAZON_URL = "https://www.flipkart.com/"

AMAZON_SMARTPHONE_PAGE_URL_1 = "https://www.flipkart.com/search?q=books&as=on&as-show=on&otracker=AS_Query_OrganicAutoSuggest_1_4_na_na_na&otracker1=AS_Query_OrganicAutoSuggest_1_4_na_na_na&as-pos=1&as-type=HISTORY&suggestionId=books&requestId=e3099506-e986-42aa-a212-960ce391c810&as-backfill=on"

AMAZON_SMARTPHONE_PAGE_URL_2 = "https://www.flipkart.com/search?q=books&as=on&as-show=on&otracker=AS_Query_OrganicAutoSuggest_1_4_na_na_na&otracker1=AS_Query_OrganicAutoSuggest_1_4_na_na_na&as-pos=1&as-type=HISTORY&suggestionId=books&requestId=e3099506-e986-42aa-a212-960ce391c810&as-backfill=on&page=2"

HEADERS = {
    # "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/126.0.0.0 Safari/537.36",
    # "Accept-Language": "en-US, en;q=0.5",
    "Accept-Language": "en-GB,en;q=0.9",
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/126.0.0.0 Safari/537.36",
}
            #### Smartphones ####

def get_product_name(product_page):
    try:
        # Outer Tag Object
        product_name = product_page.find(name="span", attrs={"class": "VU-ZEz"})

        # Inner NavigatableString Object
        title_value = product_name.text

        # Product Name as a string value
        title_string = title_value.strip()

    except AttributeError:
        title_string = ""

    return title_string

def get_product_price_in_rs(product_page):

    try:
        # Outer Tag Object
        product_price = product_page.find(name="div", attrs={"class": "Nx9bqj CxhGGd"})

        # Inner NavigatableString Object
        price_value = product_price.text

        # Product Price as a int value
        price = price_value.strip().replace(',', '').replace('.', '')
        price = int(price)

    except AttributeError:
        price = 0

    return price

def get_product_specifications(product_page):

    spec = {}
    key = []
    value = []

    try:
        # Outer Tag Object
        product_details_table = product_page.find(
            name="table", attrs={"class": "_0ZhAN9"}
        )
        trs = product_details_table.find_all(name="tr")

        for tr in trs:
            key.append(tr.th.text.strip().replace("\u200e", ''))
            value.append(tr.td.text.strip().replace("\u200e", ''))
            spec = dict(zip(key, value))

        temp_dict = spec.copy();
        spec.clear();

        for key, value in temp_dict.items():
            if key in [
                'Book',
                'Author',
                'Binding',
                'Publishing Date',
                'Publisher',
                'Edition',
                'Number of Pages',
                'Languages']:
                spec[key] = value

    except AttributeError:
        return  spec

    return spec

def get_product_description(product_page):

    product_description = ""

    try:
        # Outer Tag Object
        list_of_features = product_page.find(
            name="div", attrs={"class": "U+9u4y"}
        )

        spans = list_of_features.find_all(name="div", attrs={"class": "xFVion"})
        for span in spans:
            product_description += span.text.strip() + ". "

    except AttributeError:
        product_description = ""

    return product_description

for i in range(2):

    URL = AMAZON_SMARTPHONE_PAGE_URL_1

    amazon_url_response = requests.get(url=AMAZON_SMARTPHONE_PAGE_URL_1, headers=HEADERS)

    amazon_parsed_html = BeautifulSoup(amazon_url_response.content, "html.parser")

    # Scrap next page url
    smartphone_page_next_url = amazon_parsed_html.find(
        name="a",
        attrs={"class": "cn++Ap"},
    )

    product_urls_anchor_tags = amazon_parsed_html.find_all(
        name="a",
        attrs={
            "class": "_9QVEpD"
        },
    )

    print(amazon_url_response)

    print(product_urls_anchor_tags)

    for product in product_urls_anchor_tags:
        product_relative_url = product.get("href")
        product_absolute_url = AMAZON_URL + product_relative_url

        url_response = requests.get(url=product_absolute_url, headers=HEADERS)

        parsed_html_page = BeautifulSoup(url_response.content, "html.parser")

        # print(get_product_name(parsed_html_page))
        print(get_product_specifications(parsed_html_page))
        # print(get_product_price_in_rs(parsed_html_page))
        # print(get_product_description(parsed_html_page))

    URL = smartphone_page_next_url
```
